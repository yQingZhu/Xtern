```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("2020-XTern-DS.csv")
df.head(10)
```

```python
list(df.columns) #list(dating_data)
```

```python
df.isnull().any()
```

```python
# remove $ from 3 columns
df["Average_Cost"] = df["Average_Cost"].str.replace('$', '')
df["Average_Cost"] = df["Average_Cost"].str.replace(',', '.')
df["Minimum_Order"] = df["Minimum_Order"].str.replace('$', '')
df["Cook_Time"] = df["Cook_Time"].str.replace(' minutes', '')
df.head(10)
```

```python
df.loc[df['Average_Cost'] =='-','Average_Cost']=np.nan
df.loc[df['Minimum_Order'] =='-','Minimum_Order']=np.nan
df.loc[df['Rating'] =='-','Rating']=np.nan
df.loc[df['Votes'] =='-','Votes']=np.nan
df.loc[df['Reviews'] =='-','Reviews']=np.nan

df = df.apply(pd.to_numeric, errors='ignore')
df.head()
```

```python
df['Rating'].value_counts() 
```

For 'NEW', 'Opening Soon', data cleaning

```python
df.loc[df['Rating'] =='NEW','Rating']=np.nan
df.loc[df['Rating'] =='Opening Soon','Rating']=np.nan
df['Rating'] = df['Rating'].apply(pd.to_numeric)
```

Average Cook_Time is 36.92 mins and average Rating is 3.18
The mean of Latitude and Longitude is 39.50 and -85.50, which means the optimized FoodieX pick up zones would be around(39.50, -85.50)
